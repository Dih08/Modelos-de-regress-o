# Importando as bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importando os dados

```python
dados = pd.read_csv('Position_Salaries.csv')
X = dados.iloc[:, 1:-1].values
y = dados.iloc[:, -1].values
```

# Treinando a regressão linear do modelo em todo conjunto de dados

```python
from sklearn.linear_model import LinearRegression
regressao = LinearRegression()
regressao.fit(X, y)
```

# Treinando o modelo de Regressão Polinomial

```python
from sklearn.preprocessing import PolynomialFeatures
poli = PolynomialFeatures(degree = 4)
poliX = poli.fit_transform(X)
poli_1 = LinearRegression()
poli_1.fit(poliX,y)
```

# Visualizando os resultados da regressão Linear



```python
plt.scatter(X, y, color='red')
plt.plot(X, regressao.predict(X),color = 'blue' )
plt.title('Verdade ou blefe(Regressão Linear)')
plt.xlabel('Nivel de cargo')
plt.ylabel('Salario')
plt.show()
```

# Visualizando modelo de Regressão polinomial

```python
plt.scatter(X, y, color='red')
plt.plot(X, poli_1.predict(poli.fit_transform(X)
),color = 'blue' )
plt.title('Verdade ou blefe(Regressão Polinomial)')
plt.xlabel('Nivel de cargo')
plt.ylabel('Salario')
plt.show()
```

# Visualizando resultados da Regresssão Polinomial

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color='red')
plt.plot(X_grid, poli_1.predict(poli.fit_transform(X_grid)), color='blue')
plt.title('Verdade ou blefe(Regressão Polinomial)')
plt.xlabel('Nivel de cargo')
plt.ylabel('Salario')
plt.show()
```

## Prevendo o novo resultado com Regressão Linear
